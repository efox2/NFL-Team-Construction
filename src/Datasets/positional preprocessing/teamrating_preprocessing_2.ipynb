```python
ideal_team = {
    'Quarter Back':3,
    'Running Back':4,
    'Offensive Lineman':9,
    'Tight End':3,
    'Wide Receiver':6,
    'Defensive Lineman':9,
    'Line Backer':7,
    'Defensive Back':10,
    'Kicker':1,
    'Punter':1,
    }
print(sum(ideal_team.values()))
```

```python
df_20_players['Position']=df_20_players['Position'].replace({'WR':'Wide Receiver','CB':'Corner Back','TE':'Tight End','LB':'Line Backer','DE':'Defensive Lineman','DT':'Defensive Linemen','OL':'Offensive Linemen','QB':'Quarter Back','RB':'Running Back','K':'Kicker','P':'Punter','LS':'Long Snapper','S':'Safety',
'HB':'Half Back','MLB':'Line Backer','FS':'Safety','LE':'Left End','RE':'Right End','ROLB':'Line Backer','SS':'Safety','LOLB':'Line Backer',
'LT':'Left Tackle','RT':'Right Tackle','LG':'Left Guard','C':'Center','RG':'Right Guard','FB':'Full Back'})
df_20_players['Position']=df_20_players['Position'].replace({
 'Safety':'Defensive Back',
 'Half Back':'Running Back',
 'Left End':'Defensive Lineman',
 'Right End':'Defensive Lineman',
 'Left Tackle':'Offensive Lineman',
 'Right Tackle':'Offensive Lineman',
 'Left Guard':'Offensive Lineman',
 'Center':'Offensive Lineman',
 'Right Guard':'Offensive Lineman',
 'Full Back': 'Running Back',
 'Corner Back':'Defensive Back'})
df_20_players['Position']=df_20_players['Position'].replace({
 'Defensive Linemen':'Defensive Lineman',
 'Offensive Linemen':'Offensive Lineman',
 'DL':'Defensive Lineman'})
```
